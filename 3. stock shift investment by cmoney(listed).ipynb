```python
from bs4 import BeautifulSoup
import requests
from lxml import html
import pandas as pd
from lxml import etree
import time
import random
from fake_useragent import UserAgent
```

```python
import random
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from fake_useragent import UserAgent
```

```python
#載入上市公司列表
company_info = pd.read_csv ('/Users/linhengyu/Downloads/t187ap03_L.csv')
```

```python
company = []
name = []
shift_invest = []
```

```python
pages = list(company_info["公司代號"])
names = list(company_info["公司名稱"])
```

```python
# 設定 Chrome 瀏覽器
options = webdriver.ChromeOptions()

# 設定等待時間
MIN_WAIT_TIME = 2
MAX_WAIT_TIME = 5

for i in range(900,len(pages)):
    company.append(pages[i])
    name.append(names[i])
for p in pages[900:len(pages)]:
    url = f'https://www.cmoney.tw/finance/{p}/f00031'
    user_agent = UserAgent()
    options = Options()
    options.add_argument('--headless')
    options.add_argument(f'user-agent={user_agent.random}')
    driver = webdriver.Chrome(options=options)
    wait_time = random.uniform(MIN_WAIT_TIME, MAX_WAIT_TIME)
    time.sleep(wait_time)
    driver.get(url)
    td_list = driver.find_elements(By.CSS_SELECTOR, 'td.align-left')
    total_invest = []
    for td in td_list:
        total_invest.append(td.text)
    total_invest = list(set(total_invest))
    shift_invest.append(total_invest)
    
driver.quit()
```

```python
df = pd.DataFrame({"shift_investment": shift_invest}) 
df1 = pd.DataFrame(df['shift_investment'].to_list())
df2 = pd.DataFrame({"stock_code": company})
df3 = pd.DataFrame({"stock_name": name})
df4 = pd.concat([df3,df2,df1],axis=1)
```
